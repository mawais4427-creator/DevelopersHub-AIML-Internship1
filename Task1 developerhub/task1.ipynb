Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries imported successfully!")
```

Load the Dataset

```python
# Load Iris dataset directly from seaborn (no download needed)
df = sns.load_dataset('iris')

# Show first 5 rows
print("First 5 rows of the dataset:")
df.head()
```

Explore Shape & Columns

```python
# How many rows and columns?
print("Shape of dataset:", df.shape)

# Column names
print("Column names:", df.columns.tolist())

# Data types
print("\nData types:")
print(df.dtypes)
```

Statistical Summary

```python
# Statistical summary (min, max, mean, etc.)
print("Statistical Summary:")
df.describe()
```

Check Missing Values

```python
# Check for missing values
print("Missing values in each column:")
print(df.isnull().sum())

# Count flowers per species
print("\nNumber of flowers per species:")
print(df['species'].value_counts())
```

## Visualizations

Below are 4 visualizations created to understand
the Iris dataset better.

Scatter Plot

```python
#  Scatter Plot
plt.figure(figsize=(8, 5))

sns.scatterplot(
    data=df,
    x='sepal_length',
    y='sepal_width',
    hue='species',       # different color per species
    palette='Set1',
    s=80                 # dot size
)

plt.title('Scatter Plot: Sepal Length vs Sepal Width by Species',
          fontsize=13)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.tight_layout()
plt.show()

print("✅ Scatter plot created!")
```

Histograms

```python
#Histograms
df.drop('species', axis=1).hist(
    bins=20,
    figsize=(10, 6),
    color='steelblue',
    edgecolor='black'
)

plt.suptitle('Histogram: Distribution of All Features',
             fontsize=14, y=1.02)
plt.tight_layout()
plt.show()

print("✅ Histograms created!")
```

Box Plots

```python
# Box Plots
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

features = ['sepal_length', 'sepal_width',
            'petal_length', 'petal_width']

for i, feature in enumerate(features):
    ax = axes[i//2, i%2]
    sns.boxplot(
        data=df,
        x='species',
        y=feature,
        palette='pastel',
        ax=ax
    )
    ax.set_title(f'Box Plot of {feature}', fontsize=11)
    ax.set_xlabel('Species')
    ax.set_ylabel(feature)

plt.suptitle('Box Plots: All Features by Species',
             fontsize=14)
plt.tight_layout()
plt.show()

print("✅ Box plots created!")
```

Heatmap

```python
#Correlation Heatmap (Bonus)
plt.figure(figsize=(7, 5))

sns.heatmap(
    df.drop('species', axis=1).corr(),
    annot=True,          # show numbers
    fmt='.2f',           # 2 decimal places
    cmap='coolwarm',
    linewidths=0.5
)

plt.title('Heatmap: Correlation Between Features',
          fontsize=13)
plt.tight_layout()
plt.show()

print("✅ Heatmap created!")
```
